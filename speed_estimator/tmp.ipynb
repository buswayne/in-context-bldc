```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trapezoidal_emf(theta):
    theta = theta + 4 / 6 * np.pi
    theta = np.mod(theta, 2 * np.pi)
    if 0 <= theta < 1 * np.pi / 6:
        return 1 / (np.pi / 6) * theta
    elif np.pi / 6 <= theta < 5 * np.pi / 6:
        return 1
    elif 5 * np.pi / 6 <= theta < 7 * np.pi / 6:
        return -1 / (np.pi / 6) * (theta - np.pi)
    elif 7 * np.pi / 6 <= theta < 11 * np.pi / 6:
        return -1
    else:
        return 1 / (np.pi / 6) * (theta - 2 * np.pi)
```

```python
t = np.arange(0,10,0.01)
bemf_a = [trapezoidal_emf(t) for t in t]
bemf_b = [trapezoidal_emf(t - 2 * np.pi / 3) for t in t]
bemf_c = [trapezoidal_emf(t + 2 * np.pi / 3) for t in t]

plt.plot(t, bemf_a, label='a', linestyle='--')
plt.plot(t, bemf_b, label='b', linestyle='--')
plt.plot(t, bemf_c, label='c', linestyle='--')
plt.legend()
plt.show()
```
